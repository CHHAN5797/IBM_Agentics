```python
import csv
import re
from pathlib import Path

# CSV 파일 경로 (환경에 맞게 수정)
csv_path = Path("/Users/chunghyunhan/Projects/agentics/src/agentics/assets_registry/dao_registry.csv")

def lookup_ucid_by_space(csv_path: Path, space: str):
    if not csv_path.exists():
        raise FileNotFoundError(f"CSV not found: {csv_path}")

    def _norm(s: str) -> str:
        return re.sub(r"\s+", "", (s or "").replace("\u200b", "").strip().lower())

    target = _norm(space)
    with csv_path.open("r", encoding="utf-8") as fh:
        reader = csv.DictReader(fh)
        for row in reader:
            row_space = (row.get("space") or "").strip()
            if _norm(row_space) == target:
                return {
                    "space": row_space,
                    "CMC_ucid": row.get("CMC_ucid") or row.get("ucid"),
                    "cmc_ticker": row.get("cmc_ticker") or row.get("ticker"),
                    "defillama_name": row.get("name_defillama") or row.get("defillama_name") or row.get("name"),
                }
    return None

# === 실행 예시 ===
result = lookup_ucid_by_space(csv_path, "aavedao.eth")
print(result)
```
